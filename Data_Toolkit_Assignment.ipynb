1.  NumPy (Numerical Python) is a fundamental open-source Python library for scientific computing and data analysis. It provides a high-performance n-dimensional array object (ndarray) along with a vast collection of mathematical functions for operations such as linear algebra, Fourier transforms, statistical analysis, and random simulations.

Unlike Python lists, NumPy arrays are homogeneous (all elements have the same data type) and stored in contiguous memory, enabling operations to run up to 50x faster than native lists. This efficiency comes from vectorization (eliminating explicit loops) and broadcasting (automatic alignment of array shapes for element-wise operations).


2.  You can compute the element-wise maximum between a 1D array and a 2D array in Python using NumPy broadcasting. Broadcasting automatically expands the smaller array to match the shape of the larger one without explicit copying, making operations efficient.

Example:

import numpy as np
  #2D array
arr2d = np.array([[1, 5, 3],
                 [4, 2, 8]])
  #1D array
arr1d = np.array([3, 4, 7])
 #Element-wise maximum using broadcasting
result = np.maximum(arr2d, arr1d)
print(result)

Output:

[[3 5 7]
[4 4 8]]

Here, arr1d is broadcast across each row of arr2d because their shapes are compatible:

 - arr2d.shape → (2, 3)

 - arr1d.shape → (3,) (treated as (1, 3) for broadcasting)

NumPy compares shapes from right to left, and dimensions are compatible if they are equal or one of them is 1.

4.  The groupby() method in Pandas is a powerful tool for splitting data into groups, applying functions to each group, and combining the results.
It is commonly used for data aggregation, transformation, and analysis.

The groupby() operation follows the Split–Apply–Combine strategy:

 - Split – Divide the DataFrame into groups based on one or more keys (columns or index levels).
 - Apply – Apply a function (aggregation, transformation, or filtering) to each group.
 - Combine – Merge the results back into a DataFrame or Series.

**Example**
import pandas as pd

 #Sample Data
data = {
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],

df = pd.DataFrame(data)

 #Group by Department and calculate average salary
avg_salary = df.groupby('Department')['Salary'].mean()

6.  Here’s a clear comparison between NumPy arrays and Python lists, focusing on performance, functionality, and use cases.

 **1. Data Type Handling**

  - **Python List:**
-  Can store elements of different data types in the same list.
-  Example: [1, "hello", 3.5] is valid.

  - **NumPy Array:**

- Stores elements of the same data type (homogeneous).
- This allows for faster computation and less memory usage.

 **2. Performance**

 - **Python List:**
- Slower for numerical operations because operations are done element-by-element in Python loops.
 - **NumPy Array:**
- Much faster for numerical computations because it uses vectorized operations implemented in C.

 **3. Memory Usage**

 - **Python List:**
- Stores references to objects, so it uses more memory.
 - **NumPy Array:**
- Stores data in contiguous memory blocks, making it more memory-efficient.

 **4. Functionality**

 - **Python List:**
- General-purpose container for any type of object.
- Limited built-in mathematical operations.
 - **NumPy Array:**
- Rich set of mathematical, statistical, and linear algebra functions.
- Supports multi-dimensional arrays (matrices, tensors).

 **5. Broadcasting**

 - **Python List:**
- No broadcasting; you must loop manually for element-wise operations.
 - **NumPy Array:**
- Supports broadcasting, allowing operations between arrays of different shapes without explicit loops.

 **6. Example Comparison**

 import numpy as np

 #Python list operation

start = time.time()

8.  In NumPy, a vectorized operation means performing a mathematical or logical operation on entire arrays at once, without writing explicit Python loops.

Instead of iterating element-by-element in Python, NumPy uses highly optimized C and Fortran code under the hood to apply the operation to all elements in parallel (or in very efficient batches).

 - **No explicit loops in Python** — operations are applied to the whole array.
 - **Faster execution** — because the heavy lifting is done in compiled code.
 - **Cleaner, more readable code** — operations look like they’re applied to single numbers, but they work on arrays.

**Example: Without vs. With Vectorization**

**Non-vectorized (loop-based) approach:**

import numpy as np

arr = np.array([1, 2, 3, 4, 5])
result = np.zeros_like(arr)

for i in range(len(arr)):
    result[i] = arr[i] ** 2  # square each element

print(result)  # [ 1  4  9 16 25]

**Vectorized approach:**

import numpy as np

arr = np.array([1, 2, 3, 4, 5])
result = arr ** 2  # square all elements at once

print(result)  # [ 1  4  9 16 25]

15.   NumPy handles multidimensional arrays using its core data structure called the ndarray (N-dimensional array).
This structure is highly optimized for numerical computations and supports efficient storage, indexing, and operations on data.

**1. ndarray Structure**

  - An ndarray is a homogeneous container — all elements have the same data type.
  - The shape is a tuple indicating the size along each dimension (e.g., (3, 4, 5) means 3D array with dimensions 3×4×5).
  - The number of dimensions is called the rank or ndim.


Example:

import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.ndim)   # 2 (2D array)
print(arr.shape)  # (2, 3)


**2. Memory Layout**

  - Stored in contiguous memory blocks (row-major by default, like C).
  - Access is fast because NumPy uses strides — the number of bytes to step in each dimension.
  - Can be C-order (row-major) or Fortran-order (column-major).

**3. Indexing and Slicing**

  - Works similarly to Python lists but extended for multiple dimensions.
  - Supports:
     - Basic indexing: arr[1, 2]
     - Slicing: arr[:, 1:3]
     - Boolean indexing: arr[arr > 5]
     - Fancy indexing: arr[[0, 2], [1, 2]]


**Example**

arr = np.arange(12).reshape(3, 4)
print(arr)
 #Access element at row 1, col 2
print(arr[1, 2])  # 6
 #Slice rows and columns
print(arr[:, 1:3])

**4. Broadcasting**

  - Allows operations between arrays of different shapes without explicit loops.
  - NumPy automatically expands dimensions where possible to match shapes.

**Example:**

a = np.array([[1], [2], [3]])  # shape (3,1)
b = np.array([10, 20, 30])     # shape (3,)
print(a + b)  # Broadcasting to shape (3,3)

**5. Vectorized Operations**

  - Operations are applied element-wise without Python loops.
  - Uses optimized C code under the hood for speed.

**Example:**

arr = np.array([[1, 2], [3, 4]])
print(arr * 2)  # Element-wise multiplication

**6. Reshaping and Transposing**

  - reshape() changes the shape without changing data.
  - transpose() or .T swaps axes.
  - ravel() flattens the array.

**Example:**

arr = np.arange(6).reshape(2, 3)
print(arr.T)  # Transpose

17.  In Pandas, both apply() and map() are used to transform data, but they differ in scope, usage, and flexibility.

**1. map()**

- Scope: Works only on Series (1D data).
- Purpose: Applies a function, mapping, or dictionary to each element in the Series.
- Input: Can take

    - A function (e.g., lambda x: x*2)
    - A dictionary (for value mapping)
    - A Series (for mapping values)
- Output: Returns a new Series.

**Example:**

import pandas as pd

s = pd.Series([1, 2, 3, 4])

 #Using a function
print(s.map(lambda x: x**2))

 #Using a dictionary
print(s.map({1: 'A', 2: 'B'}))

**2. apply()**

- Scope: Works on both Series and DataFrame.
- Purpose: Applies a function along an axis (rows or columns) for DataFrames, or element-wise for Series.
- Input: A function (can be built-in, NumPy, or custom).
- Output: Can return a Series, DataFrame, or scalar depending on the function.

**Example**

print(s.apply(lambda x: x**2))

import pandas as pd

rng = pd.date_range(start='2024-01-01', periods=5, freq='D')

 #Downsample daily data to monthly mean
monthly = df.resample('M').mean()
 #Upsample to hourly, forward-fill missing values
hourly = df.resample('H').ffill()

**Rolling and Shifting** rolling() applies windowed calculations (e.g., moving averages), while shift() moves data in time without changing the index (or with freq to shift index values).

 #3-day moving average
df['rolling_mean'] = df['value'].rolling(3).mean()
 #Shift index by 2 business days
df_shifted = df.shift(2, freq='B')

tz_df = df.tz_localize('UTC').tz_convert('US/Eastern')

**Period and Timedelta Support** Use PeriodIndex for span-based data (e.g., months, quarters) and TimedeltaIndex for durations. Convert between timestamp and period with .to_period() / .to_timestamp().

 #Convert to monthly periods

21.  NumPy arrays offer several advantages over Python lists, particularly for numerical and scientific computing. These benefits stem from their design, which prioritizes performance, memory efficiency, and ease of use for mathematical operations.

**Memory Efficiency**

NumPy arrays are more memory-efficient than Python lists. While lists store elements as Python objects with additional metadata (e.g., type and reference count), NumPy arrays store elements in contiguous memory locations with a fixed data type. This reduces memory overhead significantly. For example, a Python list of 1000 elements may consume 48,000 bytes, whereas a NumPy array of the same size consumes only 8,000 bytes.

**Performance**

NumPy arrays are optimized for numerical computations. Operations like element-wise addition, multiplication, and matrix manipulations are implemented in C, making them much faster than equivalent operations on Python lists. For instance, multiplying two arrays with 1,000,000 elements each takes significantly less time with NumPy (0.0066 seconds) compared to Python lists (0.0725 seconds).

**Built-in Functionality**

NumPy provides a rich set of built-in functions for tasks like linear algebra, Fourier transforms, statistical analysis, and more. These functions are not only convenient but also highly optimized, making NumPy a powerful tool for scientific computing.

```python
# 2
import pandas as pd

def get_column_mean(df, column_name):
    """
    Calculate the mean of a specific column in a DataFrame.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the data.
        column_name (str): The name of the column to calculate the mean for.

    Returns:
        float: The mean value of the column.
    """
    # Validate if column exists
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' does not exist in the DataFrame.")

    # Ensure column is numeric
    if not pd.api.types.is_numeric_dtype(df[column_name]):
        raise TypeError(f"Column '{column_name}' is not numeric.")

    return df[column_name].mean()

if __name__ == "__main__":
    # Example DataFrame
    data = {
        'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 75000, 80000]
    }

    df = pd.DataFrame(data)
    print("DataFrame:\n", df, "\n")

    try:
        column_to_check = "Salary"  # Change this to test other columns
        mean_value = get_column_mean(df, column_to_check)
        print(f"Mean of '{column_to_check}': {mean_value:.2f}")
    except (ValueError, TypeError) as e:
        print("Error:", e)
```

```python
# 4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset: Replace with your own DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [5, 3, 6, 2, 1]
}

# Create DataFrame
df = pd.DataFrame(data)

# Step 1: Calculate the correlation matrix
corr_matrix = df.corr(method='pearson')  # 'pearson', 'spearman', or 'kendall'

# Step 2: Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,        # Show correlation values
    fmt=".2f",         # Format to 2 decimal places
    cmap='coolwarm',   # Color map
    vmin=-1, vmax=1,   # Fix scale from -1 to 1
    square=True,       # Make cells square
    linewidths=0.5,    # Line between cells
    cbar_kws={"shrink": 0.8}  # Colorbar size
)

# Step 3: Add title and show plot
plt.title("Correlation Matrix Heatmap", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 7
import numpy as np

def elementwise_multiply(arr1, arr2):
    """
    Perform element-wise multiplication of two NumPy arrays.
    Arrays must be broadcastable to the same shape.
    """
    try:
        result = np.multiply(arr1, arr2)
        return result
    except ValueError as e:
        print(f"Error: {e}")
        return None

def main():
    try:
        # Example arrays (you can modify these or take user input)
        arr1 = np.array([1, 2, 3, 4])
        arr2 = np.array([5, 6, 7, 8])

        print("Array 1:", arr1)
        print("Array 2:", arr2)

        # Perform element-wise multiplication
        result = elementwise_multiply(arr1, arr2)

        if result is not None:
            print("Element-wise multiplication result:", result)

    except Exception as e:
        print(f"Unexpected error: {e}")

if __name__ == "__main__":
    main()
```

```python
# 8
import matplotlib.pyplot as plt
import numpy as np

def plot_multiple_lines():
    try:
        # Example data
        x = np.linspace(0, 10, 100)  # 100 points from 0 to 10

        # Define multiple y values for different lines
        y1 = np.sin(x)
        y2 = np.cos(x)
        y3 = np.sin(x) * np.cos(x)

        # Create the plot
        plt.figure(figsize=(8, 5))
        plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', linewidth=2)
        plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
        plt.plot(x, y3, label='sin(x) * cos(x)', color='green', linestyle='-.', linewidth=2)

        # Add title and labels
        plt.title('Multiple Line Plot Example', fontsize=14)
        plt.xlabel('X-axis', fontsize=12)
        plt.ylabel('Y-axis', fontsize=12)

        # Add legend
        plt.legend()

        # Add grid
        plt.grid(True, linestyle='--', alpha=0.6)

        # Show the plot
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"An error occurred while plotting: {e}")

if __name__ == "__main__":
    plot_multiple_lines()
```
